```python
import os
import json
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from PIL import Image
import pydicom
import numpy as np

class MRIDataset(Dataset):
    def __init__(self, json_path, image_size=224, training=False):
        with open(json_path, 'r') as file:
            self.data_info = json.load(file)

        self.image_paths = []
        self.labels = []
        self.classes = self.data_info['class_names']
        for i, class_name in enumerate(self.classes):
            for path in self.data_info['data'][class_name]:
                self.image_paths.append(path)
                self.labels.append(i)

        self.image_size = image_size
        self.transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]) if training else transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        image = self.load_image(img_path)
        if self.transform:
            image = self.transform(image)
        return image, label

    def load_image(self, path):
        dicom = pydicom.dcmread(path)
        image = dicom.pixel_array

        # Handle multi-frame DICOM by selecting the first frame
        if image.ndim == 3 and image.shape[0] > 1:
            image = image[0]

        # Normalize the image to the range [0, 255]
        image = image.astype(float)
        image -= image.min()
        image /= image.max()
        image *= 255.0
        image = image.astype(np.uint8)

        # Convert grayscale image to RGB
        image = np.stack((image,)*3, axis=-1)

        # Convert numpy array to PIL Image
        return Image.fromarray(image, 'RGB')

# Paths to JSON configuration files
train_json_path = '/content/drive/MyDrive/Model2/DATA/configs/train_config.json'
valid_json_path = '/content/drive/MyDrive/Model2/DATA/configs/valid_config.json'
test_json_path = '/content/drive/MyDrive/Model2/DATA/configs/test_config.json'

# Create datasets
train_dataset = MRIDataset(train_json_path, training=True)
valid_dataset = MRIDataset(valid_json_path)
test_dataset = MRIDataset(test_json_path)

# Setup DataLoaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2)
val_loader = DataLoader(valid_dataset, batch_size=16, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False, num_workers=2)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet50, ResNet50_Weights
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load and modify the model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
weights = ResNet50_Weights.DEFAULT
model = resnet50(weights=weights)
for name, param in model.named_parameters():
    param.requires_grad = False  # Freeze all by default
    if "layer4" in name or "fc" in name:
        param.requires_grad = True  # Unfreeze layer4 and fc for fine-tuning

num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2)  # Assuming two classes: MCI and NC
model.to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
best_accuracy = 0.0
best_model_wts = model.state_dict()

num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    # Validation phase
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    epoch_accuracy = accuracy_score(all_labels, all_preds)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Validation Accuracy: {epoch_accuracy:.4f}')

    # Save the best model
    if epoch_accuracy > best_accuracy:
        best_accuracy = epoch_accuracy
        best_model_wts = model.state_dict().copy()
        torch.save(best_model_wts, 'best_model.pth')
        print(f"New best model saved with accuracy: {best_accuracy:.4f}")

# Optionally load the best model weights again
model.load_state_dict(torch.load('best_model.pth'))

# Test evaluation
model.eval()
all_preds = []
all_labels = []
with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

test_accuracy = accuracy_score(all_labels, all_preds)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(classification_report(all_labels, all_preds, target_names=train_dataset.classes))

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=train_dataset.classes, yticklabels=train_dataset.classes)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Class Activation Map (CAM)
def compute_cam(image, model, target_layer):
    model.eval()
    features = []
    def hook(module, input, output):
        features.append(output)

    handle = target_layer.register_forward_hook(hook)
    with torch.no_grad():
        output = model(image.unsqueeze(0).to(device))

    handle.remove()
    features = features[0].squeeze(0).cpu().numpy()

    # Weigh the feature maps using the learned weights
    weights = model.fc.weight.data.cpu().numpy()
    cam = np.dot(weights[0], features.reshape(features.shape[0], -1))
    cam = cam.reshape(features.shape[1], features.shape[2])

    # Normalize the CAM
    cam -= cam.min()
    cam /= cam.max()
    cam = np.uint8(255 * cam)

    return cam

# Example of visualizing CAM for one of the images
example_image, _ = test_dataset[0]
target_layer = model.layer4
cam = compute_cam(example_image, model, target_layer)

plt.figure(figsize=(8, 8))
plt.imshow(np.transpose(example_image.numpy(), (1, 2, 0)))
plt.imshow(cam, cmap='jet', alpha=0.5)
plt.title('Class Activation Map')
plt.axis('off')
plt.show()
```

```python
import os
import json
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet50, ResNet50_Weights
from torch.utils.data import DataLoader
from sklearn.metrics import classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import pydicom
import torchvision.transforms as transforms
from torchmetrics.classification import MulticlassConfusionMatrix, MulticlassAccuracy, MulticlassPrecision, MulticlassRecall, MulticlassF1Score


# Load and modify the model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
weights = ResNet50_Weights.DEFAULT
model = resnet50(weights=weights)
for name, param in model.named_parameters():
    param.requires_grad = False  # Freeze all by default
    if "layer4" in name or "fc" in name:
        param.requires_grad = True  # Unfreeze layer4 and fc for fine-tuning

num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2)  # Assuming two classes: MCI and NC
model.to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Function to plot training and validation loss curves
def plot_loss_curves(train_loss, val_loss, save_path='training_validation_loss.png'):
    plt.figure(figsize=(8, 6))
    plt.plot(train_loss, label='Training Loss')
    plt.plot(val_loss, label='Validation Loss')
    plt.xlabel('Epoch Number')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Training and Validation Loss per Epoch')
    plt.savefig(save_path)
    plt.show()

# Training loop
train_loss = []
val_loss = []
best_accuracy = 0.0
best_model_wts = model.state_dict()

num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    epoch_train_loss = running_loss / len(train_loader)
    train_loss.append(epoch_train_loss)

    # Validation phase
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)
            running_loss += loss.item()
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    epoch_val_loss = running_loss / len(val_loader)
    val_loss.append(epoch_val_loss)

    epoch_accuracy = accuracy_score(all_labels, all_preds)
    print(f'Epoch [{epoch+1}/{num_epochs}], Training Loss: {epoch_train_loss:.4f}, Validation Loss: {epoch_val_loss:.4f}, Validation Accuracy: {epoch_accuracy:.4f}')

    # Save the best model
    if epoch_accuracy > best_accuracy:
        best_accuracy = epoch_accuracy
        best_model_wts = model.state_dict().copy()
        torch.save(best_model_wts, 'best_model.pth')
        print(f"New best model saved with accuracy: {best_accuracy:.4f}")

# Optionally load the best model weights again
model.load_state_dict(torch.load('best_model.pth'))

# Metrics objects
num_classes = 2
conf_matrix = MulticlassConfusionMatrix(num_classes=num_classes)
accuracy = MulticlassAccuracy(num_classes=num_classes)
precision = MulticlassPrecision(num_classes=num_classes)
recall = MulticlassRecall(num_classes=num_classes)
f1_score = MulticlassF1Score(num_classes=num_classes)

# Test evaluation
model.eval()
all_preds = []
all_labels = []
with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

test_accuracy = accuracy(torch.tensor(all_preds), torch.tensor(all_labels))
test_precision = precision(torch.tensor(all_preds), torch.tensor(all_labels))
test_recall = recall(torch.tensor(all_preds), torch.tensor(all_labels))
test_f1 = f1_score(torch.tensor(all_preds), torch.tensor(all_labels))
conf_matrix_result = conf_matrix(torch.tensor(all_preds), torch.tensor(all_labels))

print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Test Recall: {test_recall:.4f}")
print(f"Test F1-Score: {test_f1:.4f}")
print(classification_report(all_labels, all_preds, target_names=train_dataset.classes))

# Confusion Matrix
sns.heatmap(conf_matrix_result.numpy(), annot=True, fmt='d', cmap='Blues', xticklabels=train_dataset.classes, yticklabels=train_dataset.classes)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Class Activation Map (CAM)
def compute_cam(image, model, target_layer):
    model.eval()
    features = []
    def hook(module, input, output):
        features.append(output)

    handle = target_layer.register_forward_hook(hook)
    with torch.no_grad():
        output = model(image.unsqueeze(0).to(device))

    handle.remove()
    features = features[0].squeeze(0).cpu().numpy()

    # Weigh the feature maps using the learned weights
    weights = model.fc.weight.data.cpu().numpy()
    cam = np.dot(weights[0], features.reshape(features.shape[0], -1))
    cam = cam.reshape(features.shape[1], features.shape[2])

    # Normalize the CAM
    cam -= cam.min()
    cam /= cam.max()
    cam = np.uint8(255 * cam)

    return cam

# Example of visualizing CAM for one of the images
example_image, _ = test_dataset[0]
target_layer = model.layer4
cam = compute_cam(example_image, model, target_layer)

plt.figure(figsize=(8, 8))
plt.imshow(np.transpose(example_image.numpy(), (1, 2, 0)))
plt.imshow(cam, cmap='jet', alpha=0.5)
plt.title('Class Activation Map')
plt.axis('off')
plt.show()

# Plot training and validation loss curves
plot_loss_curves(train_loss, val_loss)
```

```python
from fastapi import FastAPI, UploadFile, File
from fastapi.responses import JSONResponse
import torch
from torchvision.models import resnet50, ResNet50_Weights
from torchvision import transforms
from PIL import Image
import io
import numpy as np

# Load the trained model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
weights = ResNet50_Weights.DEFAULT
model = resnet50(weights=weights)
model.fc = torch.nn.Linear(model.fc.in_features, 2)  # Modify for two classes
model.load_state_dict(torch.load("best_model.pth", map_location=device))
model.to(device)
model.eval()

# Initialize FastAPI
app = FastAPI()

# Define image preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def prepare_image(image_bytes: bytes):
    image = Image.open(io.BytesIO(image_bytes)).convert("RGB")
    return transform(image).unsqueeze(0).to(device)

@app.post("/predict")
async def predict(file: UploadFile = File(...)):
    image = await file.read()
    input_tensor = prepare_image(image)
    with torch.no_grad():
        outputs = model(input_tensor)
        _, predicted = torch.max(outputs, 1)
        label = "MCI" if predicted.item() == 1 else "NC"
    return JSONResponse({"filename": file.filename, "prediction": label})

@app.get("/")
async def root():
    return {"message": "Welcome to the MRI classification API"}
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import cv2
from torchvision import transforms
from PIL import Image
import pydicom
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns

# Load the trained model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = resnet50(weights=ResNet50_Weights.DEFAULT)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2)
model.load_state_dict(torch.load('best_model.pth'))
model.to(device)
model.eval()

# Define preprocessing function
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Function to load and preprocess DICOM image
def load_dicom_image(path, image_size=224):
    dicom = pydicom.dcmread(path)
    image = dicom.pixel_array.astype(float)
    image = (np.maximum(image, 0) / image.max()) * 255.0  # Normalize the image to [0, 255]
    image = np.uint8(image)
    image = cv2.resize(image, (image_size, image_size))  # Resize image to match model input
    image = np.stack((image,) * 3, axis=-1)  # Convert to 3-channel
    return Image.fromarray(image)

# Function to compute and visualize CAM
def compute_cam(dicom_path, model, target_layer, preprocess):
    features_blobs = []

    def hook_feature(module, input, output):
        features_blobs.append(output.cpu().data.numpy())

    handle = target_layer.register_forward_hook(hook_feature)

    # Load and preprocess image
    img = load_dicom_image(dicom_path)
    tensor_img = preprocess(img).unsqueeze(0).to(device)

    # Get model predictions
    with torch.no_grad():
        outputs = model(tensor_img)
        probs = torch.nn.functional.softmax(outputs, dim=1)
        pred_class = probs.argmax().item()
        confidence = probs[0][pred_class].item()

    handle.remove()

    # Generate CAM
    features = features_blobs[0].squeeze(0)  # Shape: [2048, 7, 7]
    weight_softmax_params = model.fc.weight.cpu().data.numpy()
    cam_weights = weight_softmax_params[pred_class, :]
    cam = np.dot(features.reshape((2048, 49)).T, cam_weights).reshape(7, 7)
    cam = cv2.resize(cam, (224, 224))
    cam = np.maximum(cam, 0)
    heatmap = cv2.applyColorMap(np.uint8(255 * cam / cam.max()), cv2.COLORMAP_JET)
    heatmap[np.where(cam < 0.2)] = 0

    # Overlay CAM on the original image
    img = np.array(img)
    superimposed_img = heatmap * 0.4 + img
    plt.figure(figsize=(10, 10))
    plt.imshow(superimposed_img / 255.0)
    plt.title(f"Predicted Class: {'MCI' if pred_class == 1 else 'NC'}, Confidence: {confidence:.2f}")
    plt.axis('off')
    plt.show()

# Specify your DICOM image path and target layer
dicom_path = '/content/drive/MyDrive/Model2/DATA/Data/Test/MCI/ADNI_073_S_0445_MR_Axial_PD-T2_TSE__br_raw_20060620103704810_86_S15681_I17231(1).dcm'  # Update this path to your own DICOM file
target_layer = model.layer4  # Layer to visualize features from

# Compute and visualize CAM
compute_cam(dicom_path, model, target_layer, preprocess)
```
